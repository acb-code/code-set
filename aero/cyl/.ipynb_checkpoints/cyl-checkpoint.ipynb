It is based on the invisid theory for the pressure coefficient on a cylinder and uses Anderson, "Fundamentals of Aerodynamics," 6th ed. as a reference

First let's import the packages we need

```python
import numpy as np
import matplotlib.pyplot as plt
```

Now let's set up the variables we need defining the cylinder's surface

```python
# standard air
gamma = 1.4

# looking at angular coordinates around the full cylinder
th_max = np.pi*2.0
th = np.linspace(0,th_max,int(((th_max-0.0)/0.01)))
```

Now we'll use the invisid subsonic irrotational aerodynamic theory to calculate the pressure coefficient around the surface

```python
# non lifting
Cp_non = 1.0 - 4.0*np.sin(th)**2.0
```

```python
# lifting
R = 1.0
V = 1.0
G = 4.0*np.pi*V*R + 5.0
Cp = 1.0 - ( 4.0*np.sin(th)**2.0 + 
            ((2.0*G*np.sin(th))/(np.pi*R*V)) + (G/(2.0*np.pi*R*V))**2.0 )
```

Now let's plot the results for the pressure coefficient around the surface

```python
# plotting pressures
plt.close('all')
plt.figure()
plt.ion()
plt.title('Pressure Coefficient Over Cylinder')
plt.plot(th,Cp_non,label='Non-lifting')
plt.plot(th,Cp,label='Lifting')
plt.xlabel('Theta (rad)')
plt.ylabel('Cp')
plt.legend()
plt.show()
```

Note that $\theta$ ('Theta') starts at zero at the front center middle of the cylinder that's facing the uniform freestream
